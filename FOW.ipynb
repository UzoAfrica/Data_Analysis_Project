```python

# Create a function to convert age to ranges
def age_to_range(age):
    try:
        lower = (age // 10) * 10  # Get the lower bound of the range
        upper = lower + 9        # Get the upper bound of the range
        return f"{lower}-{upper}"  # Return the range as a string
    except:
        return None  # Handle invalid entries

# Apply the function to the column to create ranges
data['_11_Age_Range'] = data['_11_Age_of_Respondent'].apply(age_to_range)

# Save the modified data to a new variable
modified_data = data.copy()

# Save the modified data to a CSV file
# modified_data.to_csv('modified_data.csv', index=False)

# Print the modified data
modified_data.head()
```
